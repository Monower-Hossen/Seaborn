 # Bar Plor

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.barplot(x='sex',y='smoker',data = var,hue ="size")
plt.show()
```

```python
sns.barplot(x='sex',y='smoker',data = var,hue ="size",orient = "h")
plt.show()
```

```python
sns.barplot(x='sex', y='smoker',data = var, hue ="size", orient = "h", saturation =10, errcolor = "g", errwidth = 10, capsize = 0.5)
plt.show()
```

```python
sns.barplot(x='sex',y='smoker',data = var,hue ="size",orient = "h",saturation =10,capsize = 0.5)
plt.show()
```

```python
sns.barplot(x='sex',y='smoker',data = var,hue ="size",orient = "v",saturation =10, dodge = False)
plt.show()
```

```python

sns.barplot(x='sex',y='smoker',data = var,hue ="size",orient = "v",saturation =20, alpha = 0.5)
plt.show()
```

```python
sns.set (style = 'dark')
sns.barplot(x='sex',y='smoker',data = var,hue ="size",orient = "v",saturation =20)
plt.show()
```
