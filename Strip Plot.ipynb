# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day", y="total_bill",data=var,hue ="sex",palette = "rocket_r",linewidth = 1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill",data=var,hue ="sex",palette = "rocket_r",linewidth = 1,edgecolor ="r",jitter = 3)
plt.show()
```

```python

sns.stripplot(x="day", y="total_bill",data=var,hue ="sex",palette = "rocket_r",linewidth = 1,
              edgecolor ="r",jitter = 3,size = 5,marker = "<")
plt.show()
```
