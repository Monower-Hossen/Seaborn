# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.scatterplot(x = "sex", y = "size",data= var)
plt.show()
```

```python
sns.scatterplot(x = "sex", y = "size",data= var,hue="sex")
plt.show()
```

```python
sns.scatterplot(x = "sex", y = "size",data= var,hue="sex",style = "sex",size="sex",sizes=(100,40))
plt.show()
```

```python
sns.scatterplot(x = "sex", y = "size",data= var,hue="sex",style = "sex",sizes=(100,40), alpha =1 )
plt.show()
```

```python
m = ("Male": "*" ,"Female" : "o")
sns.scatterplot(x = "sex", y = "size",data= var,hue="sex",style = "sex",
                sizes=(100,40), alpha =1 )
plt.show()
```
