```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
V = sns.load_dataset('tips')
V
```

```python
sns.boxplot(y=V['total_bill'])
plt.show()
```

```python
sns.set_style('dark')
sns.barplot(x='day', y ='total_bill',data=V)
plt.show()
```

```python
sns.set_style('dark')
sns.barplot(x='day', y ='total_bill',data=V)
sns.despine()
plt.show()
```

```python
sns.set_style('white')

plt.figure(figsize=(5,12))
sns.barplot(x='day', y ='total_bill',data=V)
sns.despine()
plt.show()
```

```python
sns.set_style('white')
sns.set_context('talk',font_scale=1)
sns.barplot(x='day', y='total_bill', data=V, palette = 'cool')
plt.show()
```
