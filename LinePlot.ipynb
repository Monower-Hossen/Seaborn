# Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


var = [1,2,3,4,5,6,7]
var1 = [2,9,4,5,6,7,1]

x = pd.DataFrame({"var" : var,"var1": var1})

sns.lineplot(x="var", y = "var1",data=x)

plt.show()
```

```python
plt.plot(var,var1)
plt.show()
```

```python
e = sns.load_dataset("tips")
e
```

```python
e = sns.load_dataset("tips").head(100)
e
```

```python
sns.lineplot(x="tip", y ="smoker", data=e)

plt.show() 
```

```python
sns.lineplot(x="tip", y ="smoker", data=e,hue = "sex")

plt.show() 
```

```python
sns.lineplot(x="tip", y ="smoker", data=e,hue = "sex",size = 50)

plt.show() 
```

```python
sns.lineplot(x="tip", y ="smoker", data=e,hue = "sex",style = "sex")

plt.show() 
```

```python
sns.lineplot(x="tip", y ="smoker", data=e,hue = "sex",style = "sex",palette = "Accent",markers = ["o" , ">"], dashes = False )

plt.grid()
plt.title("Python")
plt.show() 
```
